```python
yeild_apples = [0.78, 0.81, 0.89, 0.926, 0.929, 0.931]
yeild_oranges = [0.96, 0.90, 0.87, 0.81, 0.77, 0.75]
years = range(2020, 2026)
```

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize = (10, 8))
plt.plot(years, yeild_apples, marker='>', alpha=0.5)
plt.plot(years, yeild_oranges, marker='<')
#alpha = opacity
#c = color
#ls = '-' #linestyle
#lw = linewidth = 3
#markersize ms
#markeredgewidth mew
#markeredgecolor mec
#markerfacecolor mfc (color of the filling)

plt.legend(['Apples', 'Oranges'])
plt.xlabel('year')
plt.ylabel('yeild (tons per hectare)')
plt.title('Crop yeild in Chicago')
```
